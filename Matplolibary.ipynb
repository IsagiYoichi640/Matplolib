```python
import matplotlib.pyplot as plt
```

```python
#create list of students with their names and scores
names=['Ron','Harry','Ginny','Malfoy','Hermoine','Mike','Peter',
       'Subaru','Mai','Sakuta']
scores=[75,87,80,60,95,79,50,90,100,99]
print("Data created succesfully")
```

```python
#calculate the percentage score of the student
#crete empty list to stote percentage scores
percentage_scores=[]
for x in scores:
  result = (x/100)*100
  percentage_scores.append(result)
print(percentage_scores)
```

```python
plt.plot(names,percentage_scores)
plt.title("Percentage scores of students")
plt.xlabel("Student names")
plt.ylabel("Percentage scores")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.bar(names,percentage_scores)
plt.title("Percentage scores of students")
plt.xlabel("Student names")
plt.ylabel("Percentage scores")
plt.xticks(rotation=45)
plt.show()
```

# Findings


*   From the graph, Mai is best score because she has the longest bar.
*   From the graph, Peter has the lowest score because he has the shortest bar.
* The score distribution in the classroom is quite even as there are constant length of the bars.
